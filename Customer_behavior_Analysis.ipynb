```python
import pandas as pd

df = pd.read_csv('customer_shopping_behavior.csv')
df.head()
```

Describe the numerical data alone

```python
df.describe() # numerical alone
```

Describe the category and numerical data

```python
df.describe(include = 'all') # category and numeical data
```

Check any null values present or not

```python
df.isnull().sum()
```

Fill the null values by median().Median () is more suitable than mean and It avoid the outliers

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

Convert column names into lower case and replace spaces with underscore for better coding and sql query executions.

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')

```

Rename the column names for our convenient

```python
df = df.rename(columns = {'purchase_amount_(usd)' : 'purchase_amount'} )
df.columns
```

```python
# create a column age_group

labels = ['Young Adult', 'Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'], q = 4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
df['frequency_of_purchases'].unique()
```

```python
# create purchase frequency days
frequency_map = {
    'Fortnightly' : 14 ,
    'Annually' : 365,
    'Bi-Weekly' : 14,
    'Monthly' : 30,
    'Weekly' : 7,
    'Every 3 Months' : 90,
    'Quarterly' : 90
}
df['frequency_days'] = df['frequency_of_purchases'].map(frequency_map)
```

```python
df[['frequency_of_purchases','frequency_days']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(5)

```
